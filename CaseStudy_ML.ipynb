## Standardize the data

```python
#from sklearn.preprocessing import StandardScaler
#absenteeism_scaler = StandardScaler()
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler

class CustomScaler(BaseEstimator, TransformerMixin):

    def __init__(self, columns, copy=True, with_mean=True, with_std=True):
        self.scaler = StandardScaler(copy=copy, with_mean=with_mean, with_std=with_std) #(copy, with_mean, with_std)
        self.columns = columns
        self.mean_ = None
        self.std_ = None

    def fit(self, X, y=None):
        self.scaler.fit(X[self.columns], y)
        self.mean_ = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self

    def transform(self, X, y=None, copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns, index=X.index)
        #X_not_scaled = X.loc[:, ~Xcolumns.isin(seld.columns)]
        X_not_scaled = X.drop(columns=self.columns, axis=1)
        return pd.concat([X_not_scaled, X_scaled], axis=1)[init_col_order]
```

```python
feature_name = unscaled_inputs.columns.values
```

```python
summary_tabel = pd.DataFrame(columns=['Feature name'], data=feature_name)

summary_tabel['Coefficient'] = np.transpose(reg.coef_) #We have to trasnpose the array as by default the ND arrays are rows not columns.

summary_tabel
```

```python
summary_tabel.index = summary_tabel.index + 1
summary_tabel.loc[0] = ['Intercept', reg.intercept_[0]]
summary_tabel = summary_tabel.sort_index()
summary_tabel
```

```python
summary_tabel['Odds Ratio'] = np.exp(summary_tabel['Coefficient'])
summary_tabel
```

```python
summary_tabel.sort_values('Odds Ratio', ascending=False) #by default it is in ascending order
```

###### If a coefficient is around 0 or its odds are close to 1 this means the corresponding feature is not perticularly important.
###### Because a weight of 0 implies that no matter the feature value, we will multiply it by 0.
###### For one unit change in the standardized feature, the odds increase by a multiple equal to the odds ratio(1=no change).
###### Odds X Odds Ratio = New Odds (ex-> 5:1 X 1 = 5:1(no change))

## Backward Elimination
###### The idea is that we can simplify our model by removing all features which has close to no contribution to the model.
###### When we have the p-value, we get rid off all coefficients with p-value>0.05
###### When learning with sklearn, we dont have p-value because we dont necessarily need them.
###### The reasoning is that if the weight is small enough it wont make a difference anyways.
###### If we remove these variables, the rest of our model should not really change in terms of coefficient values.
